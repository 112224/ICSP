```python
import dns.resolver
```

```python
import pandas as pd

df = pd.read_csv("dongguk_edu.txt")
```

```python
df
```

```python
tmp = df[df['port']=='None']
tmp
```

```python
hosts = tmp['domain'].to_list()
hosts
```

```python
ret = []
uncheckable = []
Resolver = dns.resolver.Resolver()
#Resolver.nameservers = ['8.8.8.8', '8.8.4.4']
qtype = ['A', 'AAAA', 'CNAME', 'MX', 'NS','SOA']
for host in hosts:
    tmp_ret = []
    for q in qtype:
        try:
            ip = Resolver.query(host,q)
            if ip:
                tmp_ret.append(ip)
                print('host = ',host,'\n', q,': ',ip[0].to_text())
        except:
            pass
    if tmp_ret:
        ret.extend(tmp_ret)
    else:
        uncheckable.append(host)
        print('host = ',host,'\n', "don't check the ip")
```
