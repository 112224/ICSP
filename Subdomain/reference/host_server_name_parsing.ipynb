```python
import pandas as pd

df = pd.read_csv("test.csv")
df[0:10]
```

```python
hosts = df.dropna(subset=['ports']).to_dict('records')
hosts
```

```python
import requests
import traceback
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)


for host in hosts:
    domain = host['domain']
    ports = host['ports']
    if ports:
        ports = ports.replace("[", "")
        ports = ports.replace("]", "")
        ports_list = ports.split(", ")
    print(domain)
    for port in ports_list:
        if port == "80":
            protocol = "http"
        elif port == "443":
            protocol = "https"
        else:
            continue
        try:
            response = requests.get(f"{protocol}://{host['domain']}", verify=False, timeout=5)
            if 'Server' in response.headers:
                print(response.headers['Server'])
        except:
            print(traceback.format_exc().splitlines()[-1])
    print(end='\n\n')
```
