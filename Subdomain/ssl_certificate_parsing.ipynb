```python
import pandas as pd

df = pd.read_csv("output.txt")
df[:10]
```

```python
hosts = df[df['port']=='443']['domain'].tolist()
hosts
```

```python
# 인증서 받아서 발급기관이랑 발급일, 만료일 출력

import time
import ssl, socket
import traceback
from cryptography import x509
from cryptography.hazmat.backends import default_backend
from cryptography.x509.oid import NameOID

for name in hosts:
    print(name)
    ctx = ssl._create_unverified_context()
    with ctx.wrap_socket(socket.socket(), server_hostname=name) as s:
        try:
            s.connect((name, 443))
            cert_der = s.getpeercert(True)
            cert = x509.load_der_x509_certificate(cert_der, default_backend())
        except:
            print(traceback.format_exc().splitlines()[-1], end='\n\n')
            continue
        if cert.issuer.get_attributes_for_oid(NameOID.COMMON_NAME):
            print(cert.issuer.get_attributes_for_oid(NameOID.COMMON_NAME)[0].value)
        else:
            print("None")
        print(cert.not_valid_before)
        print(cert.not_valid_after, end='\n\n')
        s.close()
```
