```python
import pandas as pd

df = pd.read_csv("output.txt")
df
```

```python
hosts = df[df['port']=='443']['domain'].tolist()

#hosts = df['domain'].tolist()  #timeout checking용

hosts
```

```python
import time
import ssl, socket
import traceback
import pprint

for name in hosts:
    print(name)
    ctx = ssl.create_default_context()
    with ctx.wrap_socket(socket.socket(), server_hostname=name) as s:
        s.settimeout(0.1)
        try:
            s.connect((name, 443))
            cert = s.getpeercert()
            print(cert['issuer'][-1][0][1])
            print(cert['notBefore'])
            print(cert['notAfter'], end='\n\n')
        except:
            print(traceback.format_exc().splitlines()[-1], end='\n\n')
        s.close()
```
